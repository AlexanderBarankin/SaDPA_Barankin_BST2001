### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

Методы:

Выбором Вставкой Обменом Шелла Турнирная Быстрая сортировка Пирамидальная

```python
import random
import time

m = 5
n = 5
min_limit = -250
max_limit = 1000 + 4

# Генерация матрицы с рандомными элементами
matrix = [[random.randint(min_limit, max_limit + 1) for j in range(m)] for i in range(n)]

# Вывод сгенерированной матрицы
print('Сгенирированная матрица:')
for i in range(m):
    print(matrix[i])

# Функция сортировки выбором
def selectionSort(matrix): 
    array = matrix.copy()
    for i in range(len(array) - 1):
        idMin = i
        j = i + 1
        while j < len(array):
            if array[j] < array[idMin]:
                idMin = j
            j = j + 1
        array[i], array[idMin] = array[idMin], array[i]
    return (array)

# Вывод отсортированной матрицы
start_time = time.time()
print(' ')
print('Сортировка выбором: ')
for i in range(m):
    print(selectionSort(matrix[i]))
print(' ')
print(f"{(time.time() - start_time)} секунд")


# Функция сортировки вставкой
def insertionSort(matrix):
    array = matrix.copy()
    for i in range(1, len(array)):
        temp = array[i]
        j = i - 1
        while j >= 0 and temp < array[j]:
            array[j + 1] = array[j]
            j = j - 1
        array[j + 1] = temp
    return(array)

# Вывод отсортированной матрицы
start_time = time.time()
print(' ')
print('Сортировка вставкой: ')
for i in range(m):
    print(insertionSort(matrix[i]))
print(' ')
print(f"{(time.time() - start_time)} секунд")

# Функция сортировки обменом (пузырьком)
def bubbleSort(matrix):
    array = matrix.copy()
    for i in range(len(array)):
        for j in range(0, len(array) - i - 1):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
    return(array)

# Вывод отсортированной матрицы
start_time = time.time()
print(' ')
print('Сортировка обменом (пузырьком): ')
for i in range(m):
    print(bubbleSort(matrix[i]))
print(' ')
print(f"{(time.time() - start_time)} секунд")

# Функция сортировки Шелла
def shellSort(matrix):
    array = matrix.copy()
    interval = len(array)//2
    while interval:
        for i, el in enumerate(array):
            while i >= interval and array[i - interval] > el:
                array[i] = array[i - interval]
                i -= interval
                array[i] = el
        interval = 1 if interval == 2 else int(interval * 5.0 / 11)
    return(array)
            
# Вывод отсортированной матрицы
start_time = time.time()
print(' ')
print('Сортировка Шелла: ')
for i in range(m):
    print(shellSort(matrix[i]))
print(' ')
print(f"{(time.time() - start_time)} секунд")

# Функция турнирной сортировки
def tournamentSort(matrix):
    array = matrix.copy()
    tree = [None] * 2 * (len(array) + len(array) % 2)
    index = len(tree) - len(array) - len(array) % 2
    
    for i, v in enumerate(array):
        tree[index + i] = (i, v)
    
    for j in range(len(array)):
        n = len(array)
        index = len(tree) - len(array) - len(array) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1) 
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        array[j] = x
        tree[len(tree) - len(array) - len(array) % 2 + index] = None
    return(array)

# Вывод отсортированной матрицы
start_time = time.time()
print(' ')
print('Турнирная сортировка: ')
for i in range(m):
    print(tournamentSort(matrix[i]))
print(' ')
print(f"{(time.time() - start_time)} секунд")

# Функция быстрой сортировки Хоара
def quickSort(matrix):
    array = matrix.copy()
    if len(array) < 2:
        return array
    low = []
    same = []
    high = []
    pivot = array[random.randint(0, len(array) - 1)]
    for item in array:
        if item < pivot:
            low.append(item)
        elif item == pivot:
            same.append(item)
        elif item > pivot:
            high.append(item)
    return quickSort(low) + same + quickSort(high)
    
# Вывод отсортированной матрицы
start_time = time.time()
print(' ')
print('Быстрая сортировка Хоара: ')
for i in range(m):
    print(quickSort(matrix[i]))
print(' ')
print(f"{(time.time() - start_time)} секунд")

# Функция пирамидальной сортировки
def heapify(array, heapSize, rootIndex):
    largest = rootIndex
    leftChild = (2 * rootIndex) + 1
    rightChild = (2 * rootIndex) + 2
    
    if leftChild < heapSize and array[leftChild] > array[largest]:
        largest = leftChild
        
    if rightChild < heapSize and array[rightChild] > array[largest]:
        largest = rightChild
        
    if largest != rootIndex:
        array[rootIndex], array[largest] = array[largest], array[rootIndex]

def heapSort(matrix):
    array = matrix.copy()
    n = len(array)
    for i in range(n, -1, -1):
        heapify(array, n, i)
    for i in range(n - 1, 0, -1):
        array[i], array[0] = array[0], array[i]
        heapify(array, i, 0)
    return(array)


# Вывод отсортированной матрицы
start_time = time.time()
print(' ')
print('Пирамидальная сортировка: ')
for i in range(m):
    print(heapSort(matrix[i]))
print(' ')
print(f"{(time.time() - start_time)} секунд")
    
# Встроенная функция сортировки
def integralSort(matrix):
    array = matrix.copy()
    array.sort()
    return(array)

start_time = time.time()
print(' ')
print('Встроенная функция сортировки: ')
for i in range(n):
    print(integralSort(matrix[i]))
print(' ')
print(f"{(time.time() - start_time)} секунд")
```
